# Azure ML v2 Job Submission

This notebook submits a command job that reads JSONL data from the `../data` MLTable and runs the script in `./src`.

```python
from dotenv import load_dotenv
load_dotenv(override=True)

import os
from azure.ai.ml import MLClient
from azure.identity import DefaultAzureCredential

ml_client = MLClient.from_config(
    DefaultAzureCredential(),

)
```

```python
from azure.ai.ml.entities import Environment, BuildContext

env_docker_conda = Environment(
    image="mcr.microsoft.com/azureml/curated/python-sdk-v2:42",
    conda_file="./conda.yaml",
    name="amlsdk_mltable",
    description="Environment created from a Docker image plus Conda environment.",
)
ml_client.environments.create_or_update(env_docker_conda)
```

```python
from azure.ai.ml import Input, command
from azure.ai.ml.constants import AssetTypes

job = command(
    code="./src",
    command="python job.py",
    environment="azureml://registries/azureml/environments/python-sdk-v2/versions/42",
    compute="cpucluster",
    display_name="jsonl-mltable-demo",
    description="Read JSONL data from MLTable and print a summary.",
)
```

```python
submitted_job = ml_client.jobs.create_or_update(job)
print(f"Submitted job: {submitted_job.name}")
#ml_client.jobs.stream(submitted_job.name)
```
