```python
import os
from azure.identity import DefaultAzureCredential, get_bearer_token_provider
from openai import AzureOpenAI
from dotenv import load_dotenv

load_dotenv(override=True)

    
client = AzureOpenAI(
    azure_ad_token_provider=get_bearer_token_provider(
        DefaultAzureCredential(),
        "https://cognitiveservices.azure.com/.default",
    ),
    api_version="2024-02-01",
    azure_endpoint=os.getenv("AZURE_OPENAI_AUDIO_ENDPOINT"),
)

deployment_id = os.getenv("AZURE_OPENAI_DEPLOYMENT_TRANSCRIBE_NAME")
print(deployment_id) #This will correspond to the custom name you chose for your deployment when you deployed a model."
audio_test_file = "./analysis.wav"

result = client.audio.transcriptions.create(
    file=open(audio_test_file, "rb"),            
    model=deployment_id
)

print(result)
```

```python
import os
from azure.identity import DefaultAzureCredential, get_bearer_token_provider
from openai import AzureOpenAI
import logging
from dotenv import load_dotenv
load_dotenv(override=True)

logging.basicConfig(level=logging.INFO)

client = AzureOpenAI(
    azure_ad_token_provider=get_bearer_token_provider(
        DefaultAzureCredential(),
        "https://cognitiveservices.azure.com/.default",
    ),
    api_version="2024-02-01",
    azure_endpoint=os.getenv("AZURE_OPENAI_AUDIO_ENDPOINT"),
)

deployment_id = os.getenv("AZURE_OPENAI_DEPLOYMENT_TRANSCRIBE_NAME") #This will correspond to the custom name you chose for your deployment when you deployed a model."
audio_test_file = "./dylan-song.mp3"

result = client.audio.transcriptions.create(
    file=open(audio_test_file, "rb"),            
    model=deployment_id
)

print(result)
```
