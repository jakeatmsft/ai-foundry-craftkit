```python
import os
from openai import AzureOpenAI
import logging
from dotenv import load_dotenv
load_dotenv(override=True)

logging.basicConfig(level=logging.INFO)

client = AzureOpenAI(
    api_key=os.getenv("API_KEY"),  
    api_version="2024-02-01",
    azure_endpoint = os.getenv("AZURE_OPENAI_AUDIO_ENDPOINT")
)

deployment_id = os.getenv("AZURE_OPENAI_DEPLOYMENT_TRANSCRIBE_NAME") #This will correspond to the custom name you chose for your deployment when you deployed a model."
audio_test_file = "./dylan-song.mp3"

result = client.audio.transcriptions.create(
    file=open(audio_test_file, "rb"),            
    model=deployment_id
)

print(result)
```
